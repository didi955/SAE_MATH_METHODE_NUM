```python
#Fonction aléatoire
from random import random
from math import *
import numpy as np

#Fonction de graphique et d'animation
from matplotlib.pyplot import *
from matplotlib import animation
from IPython.display import HTML
```

<div style="background-color:#81d6fe; border : 3px solid black; border-radius : 19px;margin : 0% 10% ; padding : 1%;text-align : center;font-size : 1.8em;">
    $f(x)= 2x^{2} \cos(x) -5x$
</div>

```python
def f(x):
    return 2*x**2*cos(x)-5*x
```

```python
x = np.arange(-5, 5, 0.1)
y = np.array([f(i) for i in x])

title("Représentation graphique de f(x)")
plot(x, y, label='$f(x)= 2x^{2} \cos(x) -5x$')
legend()
show()
```

**1ère Etape**: On commence par choisir un point au hasard sur la courbe

```python
x0=18
image = f(x0)
print(image)
```

```python
x = np.arange(-5, 25, 0.1)
y = np.array([f(i) for i in x])

title("Représentation graphique de f(x)")
plot(x, y, label='$f(x)= 2x^{2} \cos(x) -5x$')
plot(x0, image, marker="o", color="red")
legend()
show()
```

**2ème Etape**: On calcule la valeur de la pente soit soit la dérivée de la fonction au niveau de notre point.

```python
def df(x):
    return 4*x*cos(x)-2* x**2*sin(x)-5

pente=df(x0)
print(pente) # On a donc une pente négative à -26,2 environ
```

On avance donc dans la direction opposée à la pente avec la formule suivante: $x_{0}-\alpha * f'(x_{0})$
Il faut commencer avec une valeur de $\alpha$. <br>
Plus on prend une valeur de $\alpha$ très faible plus la descente va converger proprement et fidèlement mais va demander un plus grand nombre d'itérations et donc de temps. Dans notre cas ce n'est pas un soucis. Donc on va commencer avec une petite valeur, avec $\alpha$ = 0.02.

```python
alpha = 0.01
```

```python
def gradient(x1, nbiterations):
    for i in range(0, nbiterations):
        x1 = x1-alpha*df(x1)
    return x1
r = gradient(x0, 100)
```

```python
x = np.arange(-5, 20, 0.1)
y = np.array([f(i) for i in x])

title("Représentation graphique de f(x)")
plot(x, y, label='$f(x)= 2x^{2} \cos(x) -5x$')
plot(x0, image, marker="o", color="red")
plot(r, f(r), marker="o", color="green")
legend()
show()
```
